```python
from math import tau

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, tau, 100)

fig, (ax1, ax2) = plt.subplots(2)

ax1.set_xticks([0, tau/4, tau/2, 3*tau/4, tau])
ax1.set_yticks([-1, -1/2, 0, 1/2, 1])
ax2.set_xticks([0, tau/4, tau/2, 3*tau/4, tau])
ax2.set_yticks([-1, -1/2, 0, 1/2, 1])
# plt.grid(True)

ax1.set_xticklabels([r"$0$", r"$\tau/4$", r"$\tau/2$", r"$3\tau/4$", r"$\tau$"])
ax1.set_yticklabels([r"$-1$", r"$-1/2$", r"$0$", r"$1/2$", r"$1$"])
ax2.set_xticklabels([r"$0$", r"$\tau/4$", r"$\tau/2$", r"$3\tau/4$", r"$\tau$"])
ax2.set_yticklabels([r"$-1$", r"$-1/2$", r"$0$", r"$1/2$", r"$1$"])

ax1.set_title("One period of cosine and sine", fontsize=16)
ax1.set_xlabel(r"$\theta$", fontsize=16)
ax1.set_ylabel(r"$f(\theta)$", fontsize=16)
ax2.set_xlabel(r"$\theta$", fontsize=16)
ax2.set_ylabel(r"$f(\theta)$", fontsize=16)

ax1.annotate(r"$\cos\theta$", xy=(1.75, -0.3), xytext=(0.5, -0.75),
            arrowprops={"facecolor": "black", "width": 1}, fontsize=16)
ax1.annotate(r"$\sin\theta$", xy=(2.75, 0.5), xytext=(3.5, 0.75),
            arrowprops={"facecolor": "black", "width": 1}, fontsize=16)
ax2.annotate(r"$\cos\theta$ - $\tau/8$", xy=(2, 0.5), xytext=(3.5, 0.75),
            arrowprops={"facecolor": "black", "width": 1}, fontsize=16)

fig.set_dpi(150)
fig.subplots_adjust(hspace=0.5)

ax1.plot(x, np.cos(x), color="red", linestyle="dashed")
ax1.plot(x, np.sin(x), color="blue", linestyle="dotted")
ax2.plot(x, np.cos(x-tau/8), color="orange", linestyle="dashdot")
plt.show()
```
