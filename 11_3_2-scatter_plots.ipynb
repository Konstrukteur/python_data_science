```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
```

```python
rng = default_rng()
n_pts = 50
x = rng.standard_normal(n_pts)

y = 5*x + rng.standard_normal(n_pts)

fig, ax = plt.subplots()

ax.set_title("A Matplotlib scatter plot", fontsize=16)
ax.set_xlabel(r"x", fontsize=16)
ax.set_ylabel(r"f(\y)", fontsize=16)

ax.scatter(x, y)
plt.show()
```
